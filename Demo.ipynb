```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
ratings=pd.read_csv("dataset/toy_dataset.csv",index_col=0)
```

```python
ratings.fillna(0, inplace=True)
ratings
```

```python
def standardization(row):
    new = (row - row.mean())/(row.max()-row.min())
    return new

df_std = ratings.apply(standardization).T
print(df_std)

sparse_df = sparse.csr_matrix(df_std.values)
corrMatrix = pd.DataFrame(cosine_similarity(sparse_df),index=ratings.columns,columns=ratings.columns)
corrMatrix
```

```python
corrMatrix = ratings.corr(method='pearson')
corrMatrix.head(6)
```

```python
def similar(movie_name,rating):
    similar_sc = corrMatrix[movie_name]*(rating-2.5)
    similar_sc = similar_sc.sort_values(ascending=False)
    return similar_sc
```

```python
action = [("action1",5),("romantic2",1),("romantic3",1)]
similar_sc = pd.DataFrame()
for movie,rating in action:
    similar_sc= similar_sc.append(similar(movie,rating),ignore_index = True)

similar_sc.head(10)
```
